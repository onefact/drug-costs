```python
import pandas as pd

# Load the insulin data
data = pd.read_excel("Insulins by Country 220913.xlsx")

# Display the first few rows of the reshaped dataframe
data.head()

 # Melt the DataFrame
df_melted = pd.melt(data, id_vars=['Molecule', 'Country'], var_name='Year_Measure', value_name='Value')
 
# Split the 'Year_Measure' into 'Year' and 'Measure'
df_melted[['Year', 'Measure']] = df_melted['Year_Measure'].str.split('\n', expand=True)
 
# Drop the original 'Year_Measure' column
df_long = df_melted.drop(columns=['Year_Measure'])
 
# Pivot the DataFrame to have 'Standard Units' and 'USD MNF' in separate rows for each year
df_long = df_long.pivot_table(index=['Molecule', 'Country', 'Year'], columns='Measure', values='Value').reset_index()
 
# Reorder columns
df_long = df_long[['Molecule', 'Country', 'Year', 'Standard Units', 'USD MNF']]
 
df_long
 
# Melt the DataFrame
df_melted = pd.melt(data, id_vars=['Molecule', 'Country'], var_name='Year_Measure', value_name='Value')
 
# Split the 'Year_Measure' into 'Year' and 'Measure'
df_melted[['Year', 'Measure']] = df_melted['Year_Measure'].str.split('\n', expand=True)
 
# Drop the original 'Year_Measure' column
df_long = df_melted.drop(columns=['Year_Measure'])
 
# Pivot the DataFrame to have 'Standard Units' and 'USD MNF' in separate rows for each year
df_long = df_long.pivot_table(index=['Molecule', 'Country', 'Year'], columns='Measure', values='Value').reset_index()
 
# Reorder columns
df_long = df_long[['Molecule', 'Country', 'Year', 'Standard Units', 'USD MNF']]
 
df_long

```
